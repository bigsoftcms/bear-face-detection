```python
import dlib
import matplotlib.pyplot as plt
import cv2
```

```python
detector = dlib.cnn_face_detection_model_v1('models/bearface_network.dat')
predictor = dlib.shape_predictor('models/landmarkDetector.dat')
```

```python
img = dlib.load_rgb_image('imgs/01.jpg')
img_result = img.copy()
```

```python
dets = detector(img)

for det in dets:
    cv2.rectangle(img_result, pt1=(det.rect.left(), det.rect.top()), pt2=(det.rect.right(), det.rect.bottom()), color=255, thickness=2)
    
    shape = predictor(img, det.rect)
    
    for s in shape.parts():
        cv2.circle(img_result, center=(s.x, s.y), radius=5, color=255, thickness=-1)

plt.figure(figsize=(16, 10))
plt.imshow(img_result)
```
